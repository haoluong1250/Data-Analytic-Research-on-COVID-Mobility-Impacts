## **2. Get the raw descriptive statistics**

We will display these raw descriptive statistics in the analysis section. 

A. For Alameda: 

```python
# Generate basic statistics
alameda_basics = alameda_data.describe()

# Drop some columns
alameda_basics = alameda_basics[alameda_basics.columns[~alameda_basics.columns.isin(['Unnamed: 0', 'level_0', 'index'])]]

# Display
#alameda_basics
```

B. For Merced:

```python
# Generate basic statistics
merced_basics = merced_data.describe()

# Drop some columns
merced_basics = merced_basics[merced_basics.columns[~merced_basics.columns.isin(['Unnamed: 0', 'level_0', 'index'])]]

# Display
#merced_basics
```

__________________________________

## 3. Get the normalized descriptive statistics

Since Merced's and Alameda's populations are different, with 287,420 people for Merced, and 1,685,886 people for Alameda, we want to normalize the data to compare proportions of people rather than raw numbers. 

We will keep this data in the variables `alameda_basics_n` and `merced_basics_n`. We will display these statistics later in the analysis portion. 

A. Alameda: 

```python
# Normalize the data by population
alameda_basics_n = alameda_basics/1685886.0

# Drop Population column
alameda_basics_n = alameda_basics_n[alameda_basics_n.columns[~alameda_basics_n.columns.isin(['population'])]]

# Display
#alameda_basics_n
```

B. Merced: 

```python
# Normalize the data by population
merced_basics_n = merced_basics/287420.0

# Drop Population
merced_basics_n = merced_basics_n[merced_basics.columns[~merced_basics_n.columns.isin(['population'])]]

# Display
#merced_basics_n
```

_____________

## **4. Get comparison tables**

Generate tables to compare the descriptive statistics for Alameda and Merced side by side. 

Keep these tables in the variables `comparison_table_mortality` and `comparison_table_mortality_n`. We will display these tables later in the analysis portion. 

```python
# Compare raw, non-normalized datasets
comparison_table_mortality = alameda_basics.compare(merced_basics, 
                       keep_equal=True, 
                       keep_shape=True,
                       result_names=('alameda', 'merced'))
```

```python
# Compare the normalized datasets
comparison_table_mortality_n = alameda_basics_n.compare(merced_basics_n, 
                       keep_equal=True, 
                       keep_shape=True,
                       result_names=('alameda', 'merced'))
```
